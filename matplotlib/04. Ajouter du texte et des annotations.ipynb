# Ajouter du texte et des annotations

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

> Il est possible d'ajouter du texte aux graphique en indiquant les coordonnées où l'on veut afficher le début de celui-ci, grâce à la commande `plt.text`.
>
> Exemple : Utilisation de la commande `plt.text`:
> 
> `plt.text(3, 5,'monTexte')` affiche `monTexte` horizontalement à partir du point $(3,5)$ du graphique.

- Afficher la courbe de la fonction  $f(x)=\cos(2\pi x)$ sur l'intervalle $[0,2]$
- Délimiter l'axe des ordonnées à $[-2, 2]$
- Ajouter le texte `maximum local`, au point $(0.8,1.1)$

```python
x = np.linspace(0,2,100)
plt.plot(x, np.cos(2*np.pi*x), 'b-')
plt.ylim(-2,2)
plt.text(0.8,1.1,'Maximum Local')
plt.show()
```

> Pour ajouter une annotation avec une flèche descriptive pointant vers un point précis du graphique, on utilise `plt.annotate`.
> 
> Cette méthode prend les arguments suivants :
- le texte que l'on veut afficher.
- `xy`, qui indique les coordonnées où se trouve le point à annoter.
- `xytext`, qui indique les coordonnées du point où démarre le texte
- `arrowprops`, qui sont les propriétés de la flèche d'annotation entre accolades : couleurs, taille de la flèches, style de flèche, etc...
>
> Exemple : Utilisation de la commande `plt.annotate` :
>
> `plt.annotate('Limite', xy=(1,2), xytext=(1,2.5), arrowprops={'facecolor':'blue'})` 
>
> affiche une flèche bleue pointant ver le point de coordonnées (1, 2) et affiche le texte 'Limite' au point (1, 2.5).

- Dans un graphique quadrillé, afficher la courbe de la fonction  $f(x)=\sin(2\pi x)~\exp(-x)$  sur l'intervalle $[0,5]$
- Afficher les courbes des fonctions $\exp(-x)$ et $-\exp(-x)$  sur le même intervalle
- Rajouter une flèche d'annotation, de couleur verte, au point $(1.5,0.25)$, avec le texte `Asymptote: exp(-x)` démarrant au point $(2,0.5)$
- Rajouter une flèche d'annotation, de couleur rouge, au point $(1.5,-0.25)$, avec le texte `Asymptote: - exp(-x)` démarrant au point $(2,-0.5)$

```python
def f(x):
    return np.sin(2*np.pi*x)*np.exp(-x)

x = np.linspace(0, 5, 1000)

plt.grid(True)
plt.plot(x, f(x), 'g')
plt.plot(x,  np.exp(-x), 'b-')
plt.plot(x, -np.exp(-x), 'b-')
plt.annotate('Asymptote:  exp(-x)', (1.5, 0.25), (2, 0.5), arrowprops={'facecolor':'green'})
plt.annotate('Asymptote: -exp(-x)', (1.5,-0.25), (2,-0.5), arrowprops={'facecolor':'red'})

plt.show()
```

## Résumé de toutes les compétences :

```python
x = np.linspace(-5,2,1000)
x_sub = np.linspace(-3,-2,100)

plt.grid(True, color='grey', linestyle=':')
plt.title('Mon graphique bilan')
plt.plot(x, x**2, 'b', label='$x^2$')
plt.plot(x, np.exp(-x), 'g--', label='$\exp(-x)$')
plt.xlim(-5,2)
plt.ylim(0,25)
plt.legend(loc='upper right')
plt.annotate('Zoom', (-1.9, 7.5), (-0.5, 7.5), arrowprops={'facecolor':'red'})

plt.plot([-3, -2, -2, -3, -3],[5, 5, 10, 10, 5],'r', alpha = 0.6)

plt.axes([0.55, 0.4, 0.2, 0.2])
plt.plot(x_sub, np.exp(-x_sub), 'g--')
plt.plot(x_sub, x_sub**2, 'b')
plt.xlim(-3,-2)
plt.ylim(5,10)
plt.xticks([])
plt.yticks([])

plt.show()
```
