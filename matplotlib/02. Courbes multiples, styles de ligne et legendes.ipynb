# Courbes multiples, styles de ligne et légendes

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
````

> Pour superposer plusieurs courbes sur le même graphique, on peut cumuler les instructions plot les unes après les autres, comme ceci :
```
plt.plot(x,y)
plt.plot(u,v)
plt.plot(w,z)
```
> ou simplement utiliser la syntaxe suivante:
```
plt.plot(x,y,u,v,w,z)
```
- Créer une liste appelée $t$ composée de la séquence de $0$ à $5$, avec un pas de $0.2$, grâce à la fonction `np.arange`
- Dans un même graphe, tracer les courbes des fonctions  $t$ ,  $t^2$  et  $t^3$  en délimitant l'axe des ordonnées entre $0$ et $50$

````python
t = np.arange(0,5.2,0.2)
plt.plot(t, t, t, t**2, t, t**3)
plt.ylim([0,50]);
````

> Pour chaque paire $(x,y)$ d'arguments, il y a un troisième argument optionnel qui est une chaîne de caractères qui indique le style et la couleur de la ligne de la courbe. Les styles de courbes possibles sont:
```
'-' : ligne continue
'--' : tirets
':' : ligne en pointillé
'-.' : tirets points
```
> L'ensemble des couleurs possibles est:
```
'b' (bleu)
'g' (vert) 
'r' (rouge) 
'c' (cyan) 
'm' (magenta) 
'y' (jaune) 
'k' (noir) 
'w' (blanc)
```
> Par exemple, le format par défaut est `b-`, qui est une ligne continue bleue.

- Tracer à nouveau les 3 courbes de la question précédente en les affichant respectivement: en pointillés rouge, en ligne continue verte et en tirets bleus

````python
plt.plot(t, t, 'r:', t, t**2, 'g-', t, t**3, 'b--')
plt.ylim([0,50]);
````

> Ainsi, la commande `plt.plot(x , y , "r--", linewidth=5, marker= 'o')` affiche une ligne en tirets rouge, d'épaisseur 5, avec des marqueurs ronds.

> Créer des listes de styles de courbe et de symboles permet d'utiliser une boucle pour afficher plusieurs lignes différentes rapidement.

- Créer la liste `styles` comprenant les quatre styles de lignes possibles et la liste markers contenant les 8 symboles suivants `+oh^psxv`
- Créer une séquence $x$ de $20$ nombres entre $0$ et $100$, grâce à la méthode `linspace` de numpy
- Grâce à une boucle `for`, pour chaque  $i$  de $0$ à $7$, tracer la courbe de la fonction  $f(x)=x+(ix/5)+i$, avec le label `Ligne (i+1)`, le marqueur $i$ de la liste `markers`, le style `i%len(style)` de la liste `styles` (argument `linestyle`), et une épaisseur de ligne égale à $2$
- Ajouter une grille au graphique
- Afficher les légendes au meilleur endroit du graphique (`loc=best`)
- Ajouter le titre `8 Plots Simples`

````python
styles = ['-', '--', ':', '-.']
markers = ['+', 'o', 'h', '^', 'p', 's', 'x', 'v']
x = np.linspace(0,100,20)

def f(x,i):
    return x+(x/5.*i)+i

for i in range(8):
    plt.plot(x,f(x,i), label=f'Ligne {i+1}', marker=markers[i], linestyle=styles[i%len(styles)], linewidth=2)

plt.grid(True)
plt.legend(loc='best')
plt.title('8 Plots Simples');
````
