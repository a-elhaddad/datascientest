# Boîtes à moustaches (Box plots) et camemberts (Pie plots)

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
````

> La commande `plt.boxplot` permet d'afficher des boites à moustaches sur un graphique. Elle prend en argument une séquence de séquences de valeurs.
>
> La boîte donne les limites du premier et dernier quartile, la barre du milieu donne la médiane.
>
> Les moustaches vont jusqu'à la valeur la plus extrême dans la limite de 1.5 fois la hauteur de la boîte, et les points au-delà des moustaches sont représentés par des +.
>
> Exemple :

````python
plt.boxplot([[1, 2, 3, 4, 5], [7, 5, 8, 4, 9, 5, 7]]);
````

````python
x = [10, 12, 13, 20, 25]
y = [13, 16, 28, 38]
plt.boxplot([x,y])
plt.show()
````

> Les camemberts sont des diagrammes circulaires divisés par secteurs (wedges). C'est une manière efficace de représenter l'information lorsque l'on s'intéresse principalement à la comparaison d'un secteur avec le camembert tout entier plutôt qu'entre deux secteurs.
>
> La méthode `pie` permet de tracer des diagrammes circulaires à partir d'une séquence X.
> 
> Chaque élément  $x_i$  de X est représenté par une part proportionnelle à  $\dfrac{x_i}{\sum_i x_i}$. Si la somme des $x_i$ est inférieure à 1, le camembert est représenté à l'aide des valeurs de X directement, aucune normalisation n'est effectuée, ce qui laisse un espace vide.

- Créer une figure aux dimensions `(6, 6)`.
- Créer une liste `x` comprenant les nombres `[1, 2, 3, 4, 10]`.
- Afficher un diagramme circulaire à partir de `x`, et lui donner les labels `['A', 'B', 'C', 'D', 'E']`.
- Afficher les légendes sur le graphique.

````python
plt.figure(figsize=(6,6))
plt.pie([1, 2, 3, 4, 10], labels=['A', 'B', 'C', 'D', 'E'])
plt.legend()
plt.show()
````

> De nombreux paramètre permettent de customiser son camembert.
> 
> `explode`: tableau de la même taille que la séquence de données, permet d'éloigner une ou plusieurs part du centre en indiquant de quelle fraction de rayon chaque part doit être éloignée (0 par défaut).
> 
> `colors`: séquence de couleurs à utiliser pour les parts.
> 
> `labeldistance` : la distance des labels au centre (> 1 pour être à l'extérieur du cercle).
>
> `autopct` : une fonction (lambda) qui prend le pourcentage de la part et renvoie ce qui doit être affiché pour ce pourcentage.
>
> `pctdistance` : la distance au centre à laquelle le pourcentage précédent doit être affiché (1 = sur le cercle).
>
> `shadow=True` : indique qu'il faut afficher une ombre.

- Créer une figure de taille `(7, 7)`.
- Créer la liste `x=[2, 3, 4, 5, 6, 9]`.
- Afficher un diagramme circulaire à partir des données de `x`, avec les labels de `'A'` à `'F'` et les couleurs suivantes : rouge, orange, jaune, vert, bleu et violet.
- Éloigner la quatrième part du centre de `0.2`.
- Afficher le pourcentage de chaque part, arrondi à 2 chiffres après la virgule , suivi du signe `'%'`.
- Afficher ces pourcentages à une distance de 0.7 du centre, et les labels à une distance de 1.2.
- Rajouter des ombres au camembert.
- Afficher les légendes sur le graphique.

````python
plt.figure(figsize=(7,7))
x = [2, 3, 4, 5, 6, 9]
labels = ['A', 'B', 'C', 'D', 'E', 'F']
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'violet']
explode = [0,0,0,0.2,0,0]

plt.pie(x, labels=labels, colors=colors, explode=explode, shadow=True, autopct='%.2f%%', pctdistance=0.7, labeldistance=1.2)
plt.legend()

plt.show()
````
