# Tracer des courbes avec matplotlib

> Matplotlib est une bibliothèque de Python qui constitue un puissant outil pour tracer et visualiser des données. Elle est destinée à produire des graphiques de toutes sortes (voir ici pour une galerie d'images produites avec Matplotlib)
> 
> Matplotlib contient la sous-bibliothèque pyplot qui crée une interface proche de celle du logiciel commercial Matlab et qui contient des fonctions très similaires à celui-ci.
>
> Il existe des bibliothèques pour améliorer l'affichage des figures, ou leur donner un style différent, comme Seaborn, mais nous ne les intégrerons pas à cette formation.
>
> Par ailleurs, Matplotlib est une bibliothèque très riche, toutes ses fonctionnalités ne peuvent être traitées. Cette formation fait le choix d'explorer certaines fonctions plus que d'autres, l'objectif général étant de rendre un datascientist autonome sur le module en fin de formation.

> Pour commencer, il faut importer le module matplotlib.pyplot sous le nom raccourci de `plt`. Une fois un graphique construit, c'est l'instruction `plt.show` qui permettra de le visualiser.
>
> Cependant, sur un Notebook, comme ici sur Jupyter, rajouter en début de page `%matplotlib inline` permet d'afficher automatiquement les figures à chaque modification ou utilisation d'une commande pyplot, après exécution de la cellule. Souvent, on utilisera aussi la bibliothèque numpy.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

> Une courbe est par définition un ensemble de points de coordonnées $(x, y)$ reliés ou pas par une ligne brisée. Plus le nombre de points est important, plus la courbe aura un aspect lisse.
> 
> La méthode plot permet de tracer des courbes qui relient des points dont les abscisses et ordonnées sont fournies dans des listes ou tableaux.
>
> Pour tracer un graphe avec les valeurs $x$ en abscisses et les valeurs $y$ en ordonnées on écrit: `plt.plot(x,y)`.

- Tracer une courbe prenant en abscisses les valeurs $[0,2,4,6]$ et en ordonnées les valeurs $[1,4,4,8]$

```python
## Insérez votre code ici
x = [0,2,4,6]
y = [1,4,4,8]
plt.plot(x,y);
```

> Si l'on n'insère qu'une seule liste ou tableau dans la commande `plot`, matplotlib présume que c'est une séquence des valeurs des ordonnées $y$ et génère les valeurs des abscisses automatiquement. Les valeurs des abscisses seront les positions des valeurs de $y$ à partir de $0$.

- Tracer une courbe à partir de la liste $[1,3,2,4]$

```python
plt.plot([1,2,3,4]);
```

> Pour ajouter un titre aux graphes on utilise la méthode `title`.
>
> Pour ajouter des labels aux axes on utilise les méthodes `xlabel` et `ylabel`.

- Tracer une courbe passant par les points suivants: (50,1) , (140,2) , (200,4).
- Donner à la figure le titre 'Le titre de ma courbe'
- Donner respectivement à l'axe des abscisses et l'axe des ordonnées les labels 'abscisses' et 'ordonnees'

```python
x = [50,140,200]
y = [1,2,4]
plt.plot(x,y)
plt.title('Le titre de ma courbe')
plt.xlabel('Abscisses')
plt.ylabel('Ordonnees');
```

> Il est possible de délimiter les domaines des abscisses et des ordonnées en utilisant la méthode axis.
>
> Elle s'utilise ainsi : `axis([xmin, xmax, ymin, ymax])`.
>
> Pour fixer indépendamment le domaines des abscisses ou des ordonnées on peut utiliser respectivement `xlim([xmin, xmax])` ou `ylim([ymin, ymax])`.

- Tracer une courbe passant par les points suivants: $(50,1)$ , $(140,2)$ , $(200,4)$
- Encadrer le graphique entre les abscisses $80$ et $180$ et les ordonnées de $1$ à $5$

```python
x = [50,140,200]
y = [1,2,4]
plt.plot(x,y)
plt.axis([80,180,1,5]);
```

> La fonction `plot` ne fait que relier des points dans l'ordre que l'on lui donne, il est possible de lui fournir plusieurs points avec la même abscisse, pour tracer une forme.

- Créer les listes $x$ et $y$ suivantes: $x = [0, 0, 1, 1, 0, 0.5, 1]$ , $y = [1, 0, 0, 1, 1, 2, 1]$
- Relier ces points grâce à la fonction `plot`, et délimiter les deux axes entre $-1$ et $2$

```python
x = [0, 0, 1, 1, 0, 0.5, 1] 
y = [1, 0, 0, 1, 1, 2, 1]
plt.plot(x,y)
plt.axis([-1,2,-1,2]);
```

> De la même façon, on peut tracer une courbe paramétrée, à partir d'une séquence $t$ en donnant à la méthode `plot` une fonction de $t$ en abscisse et une autre en ordonnée.

- Créer grâce à la méthode `linspace` de Numpy, une séquence de $100$ nombres entre $0$ et $2\pi$.
- Tracer la courbe paramétrée définie par  $f(t)=(sin(2t),sin(3t))$ pour  $t\in[0,2\pi]$

```python
t = np.linspace(0,2*np.pi,100)

plt.plot(np.sin(2*t), np.sin(3*t));
```
