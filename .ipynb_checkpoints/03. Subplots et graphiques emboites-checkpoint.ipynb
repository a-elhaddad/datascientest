# Subplots et graphiques emboîtés

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

> Il est possible d'afficher plusieurs graphiques sur la même figure, grâce à la commande `subplot` qui crée une figure automatiquement et la divise autant qu'on le souhaite.
>
> La commande `subplot` prend en arguments: le nombre de lignes de la figure (`numrows`) , le nombre de colonnes (`numcols`) et le numéro du graphique sur lequel on souhaite travailler (compris entre $1$ et ${\rm numrows} \times {\rm numcols}$).
>
> Les virgules sont optionnelles si ${\rm numrows} \times {\rm numcols} < 10$ . Ainsi `subplot(211)` est identique à `subplot(2,1,1)`.

Exemple : Ecrire `plt.subplot(3,2,2)` en début de cellule crée et partage une figure en $3\times 2$ emplacements de graphes ($3$ lignes et $2$ colonnes) et sélectionne le 2ème emplacement pour les instructions graphiques suivantes.

- Créer, grâce à la méthode `linspace` du package numpy, un tableau 1D appelé `x` de $400$ valeurs allant de $0$ à $2\pi$ 
- Afficher sur une même figure les quatre graphiques suivant :
 - $f(x)=sin(x^2)$
 - $f(x)=cos(x^2)$
 - $f(x)=sin^2(x)$
 - $f(x)=cos^2(x)$ et $f(x)=sin^2(x)$


```python
x = np.linspace(0, 2*np.pi, 400)
functions = [np.sin(x**2), np.cos(x**2), np.cos(x)**2, np.sin(x)**2]
styles = ['r-', 'g-.', 'b--', 'y:']

for iplot in range(1,5):
    plt.subplot(2,2,iplot)
    plt.plot(x, functions[iplot-1], styles[iplot-1])
    if iplot == 4:
        plt.plot(x, functions[iplot-2], styles[iplot-2])
    
plt.show()
```

> Sur matplotlib, il est facile d'insérer un graphe dans un autre grâce à la méthode `axes`. Il suffit de lui donner en argument une liste comprenant les positions $x$ et $y$ du point en bas à gauche de ce graphe, avec des valeurs entre $0$ et $1$ (indépendamment des axes choisis), et les dimensions largeur et hauteur, entre $0$ et $1$.
>
> Pour plus de visibilité, il est recommandé de donner à cet axe un fond de couleur différente, grâce à l'ajout de l'argument `facecolor`. Par exemple, `plt.axes([.5, 0.6, .2, .2], facecolor='grey')` crée un graphique avec fond gris dont le point en bas à gauche se situe en abscisse au milieu du graphique actuel, en ordonnée à $60$% de hauteur, et dont la largeur et la hauteur représentent $20$% de celles du graphique actuel.

- Créer une séquence de $40$ valeurs entre $0.1$ et $5$, appelé `x`
- Dans un graphe, tracer la courbe correspondant à la fonction $\exp(x)$ en vert, et lui donner le titre `Fonction exp`
- Dans un sous graphe, tracez la courbe de la fonction $\log(x)$ sur fond gris clair (`lightgrey`), à la position $(0.6,0.6)$ du graphe précédent, de taille $(0.2,0.2)$. Supprimer les axes et lui donner le titre `Fonction log`

```python
x = np.linspace(0.1, 5, 40)

plt.plot(x, np.exp(x), 'g')
plt.title('Fonction exp')

plt.axes([0.6, 0.6, 0.2, 0.2], facecolor='lightgrey')
plt.plot(x, np.log(x))
plt.xticks([])
plt.yticks([])
plt.title('Fonction log')

plt.show()
```
